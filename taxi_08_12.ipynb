```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import nltk
import re
import seaborn as sns
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
from sklearn import cross_validation
from sklearn.linear_model import LogisticRegression
```

```python
taxi_df = pd.read_csv('taxi_08_12_churn.csv', ";")
```

```python
taxi_df = taxi_df.drop(taxi_df.columns[0], axis=1)
```

```python
taxi_df = taxi_df.fillna(0)
```

```python
# Compute average value of all average prices for this period.
average_price_for_period = taxi_df[["average_price_01_2014", "average_price_02_2014", "average_price_03_2014", "average_price_04_2014", "average_price_05_2014"]].mean(axis=1)
```

```python
# Segregate valuesю
taxi_df["average_price_for_period"] = average_price_for_period
customer_churn = taxi_df["average_price_for_period"][taxi_df["churn"] == 0]
customer_not_churn = taxi_df["average_price_for_period"][taxi_df["churn"] == 1]
```

```python
# Plot
customer_churn.plot(kind='hist', figsize=(15,15), bins=100, xlim=(0,40))
```

```python
customer_not_churn.plot(kind='hist', figsize=(15,10), bins=100, xlim=(0,200))
```

```python
taxi_df['average_price_for_period'] = taxi_df['average_price_for_period'].astype(int)
average_price = taxi_df[["average_price_for_period", "churn"]].groupby(['average_price_for_period'],as_index=False).mean()
sns.barplot(x='average_price_for_period', y='churn', data=average_price)
```

```python
# Compute the number of cancellation by different reasons.
sum_canceled_price = taxi_df[["canceled_price_01_2014", "canceled_price_02_2014", "canceled_price_03_2014", "canceled_price_04_2014", "canceled_price_05_2014"]].sum(axis=1)
sum_canceled_no_wait = taxi_df[["canceled_no_wait_01_2014", "canceled_no_wait_02_2014", "canceled_no_wait_03_2014", "canceled_no_wait_04_2014", "canceled_no_wait_05_2014"]].sum(axis=1)
sum_canceled_no_more = taxi_df[["canceled_no_more_01_2014", "canceled_no_more_02_2014", "canceled_no_more_03_2014", "canceled_no_more_04_2014", "canceled_no_more_05_2014"]].sum(axis=1)
sum_canceled_no_car = taxi_df[["canceled_no_car_01_2014", "canceled_no_car_02_2014", "canceled_no_car_03_2014", "canceled_no_car_04_2014", "canceled_no_car_05_2014"]].sum(axis=1)
```

```python
taxi_df["sum_canceled_price"] = sum_canceled_price
taxi_df["sum_canceled_no_wait"] = sum_canceled_no_wait
taxi_df["sum_canceled_no_more"] = sum_canceled_no_more
taxi_df["sum_canceled_no_car"] = sum_canceled_no_car
```

```python
# Plots for all reasons.
taxi_df['sum_canceled_price'] = taxi_df['sum_canceled_price'].astype(int)
data_canceled_price = taxi_df[["sum_canceled_price", "churn"]].groupby(['sum_canceled_price'],as_index=False).mean()
sns.barplot(x='sum_canceled_price', y='churn', data=data_canceled_price)
```

```python
taxi_df['sum_canceled_no_wait'] = taxi_df['sum_canceled_no_wait'].astype(int)
data_canceled_price = taxi_df[["sum_canceled_no_wait", "churn"]].groupby(['sum_canceled_no_wait'],as_index=False).mean()
sns.barplot(x='sum_canceled_no_wait', y='churn', data=data_canceled_price)
```

```python
taxi_df['sum_canceled_no_more'] = taxi_df['sum_canceled_no_more'].astype(int)
data_canceled_price = taxi_df[["sum_canceled_no_more", "churn"]].groupby(['sum_canceled_no_more'],as_index=False).mean()
sns.barplot(x='sum_canceled_no_more', y='churn', data=data_canceled_price)
```

```python
taxi_df['sum_canceled_no_car'] = taxi_df['sum_canceled_no_car'].astype(int)
data_canceled_price = taxi_df[["sum_canceled_no_car", "churn"]].groupby(['sum_canceled_no_car'],as_index=False).mean()
sns.barplot(x='sum_canceled_no_car', y='churn', data=data_canceled_price)
```

```python
# Calculate the number of trips for different class.
sum_done_count_econome = taxi_df[["done_count_econome_01_2014", "done_count_econome_02_2014", "done_count_econome_03_2014", "done_count_econome_04_2014", "done_count_econome_05_2014"]].sum(axis=1)
sum_done_count_standart = taxi_df[["done_count_standart_01_2014", "done_count_standart_02_2014", "done_count_standart_03_2014", "done_count_standart_04_2014", "done_count_standart_05_2014"]].sum(axis=1)
sum_done_count_comfort = taxi_df[["done_count_comfort_01_2014", "done_count_comfort_02_2014", "done_count_comfort_03_2014", "done_count_comfort_04_2014", "done_count_comfort_05_2014"]].sum(axis=1)
```

```python
taxi_df["sum_done_count_econome"] = sum_done_count_econome
taxi_df["sum_done_count_standart"] = sum_done_count_standart
taxi_df["sum_done_count_comfort"] = sum_done_count_comfort
```

```python
# Find clients who had no trips for this period.
indexis = []
for index, i in enumerate(taxi_df[['sum_done_count_econome', 'sum_done_count_standart', 'sum_done_count_comfort']].apply(max, axis=1)):
    if not i:
        indexis.append(index)
```

```python
# Get set without such clients.
taxi_order_df = taxi_df.drop(taxi_df.index[indexis])
```

```python
# Create columns with type of the greatest number of trip.
taxi_order_df["type_order"] = taxi_order_df[['sum_done_count_econome', 'sum_done_count_standart', 'sum_done_count_comfort']].idxmax(axis=1)
```

```python
# Plots.
sns.countplot(x='churn', hue="type_order", data=taxi_order_df, order=[1,0])
```

```python
# Find set for using in cross validation.
taxi_data = taxi_df.drop(taxi_df.columns[:-8], 1)
taxi_data_answer = taxi_df.drop(taxi_df.columns[-9:], 1)
taxi_data_answer = taxi_data_answer.drop(taxi_data_answer.columns[:-1], 1)
```

```python
# Compute random split into training and test sets.
X_train, X_test, y_train, y_test = cross_validation.train_test_split(taxi_data, taxi_data_answer, test_size=0.4, random_state=0)
```

```python
# Using a logistic regression.
logreg = LogisticRegression(C=1)
logreg.fit(X_train, y_train)
```

```python
Y_pred = logreg.predict(X_test)
logreg.score(X_train, y_train)
```

```python
# get Correlation Coefficient for each feature using Logistic Regression
coeff_df = DataFrame(taxi_data.columns.delete(0))
coeff_df.columns = ['Features']
coeff_df["Coefficient Estimate"] = pd.Series(logreg.coef_[0])

# preview
coeff_df
```
